#EJERCICIO: Ondas en un estanque

Gráfica de la onda generada al caer una piedra

```python
from numpy import zeros, array, loadtxt
from math import pi, sqrt, sin
from pylab import imshow, xlabel,ylabel, plot, show
puntos=500
dimen=100. #cm
paso=dimen/puntos #cm/puntos
lmda=5.0 #cm
k=2*pi/lmda
zita=1.0 #cm
x1=float(input("Ingrese la posición en x: "))
y1=float(input("Ingrese la posición en y: "))

x=zeros(puntos, float)
y=zeros(puntos, float)

for i in range(puntos):
    x[i]=paso*i #creo vectores
    y[i]=paso*i
    
ondas=open("ondas_estanque.dat", "w")

for xx in range( len(x) ):
    for yy in range( len(y) ):
        r = sqrt((x1-x[xx])*(x1-x[xx]) + (y1-y[yy])*(y1-y[yy]) )
        ondas.write("%0.1f " % (zita*sin(k*r) ))
        #ondas.write("%0.1f " % (sqrt( (x1-x[xx])*(x1-x[xx]) + (y1-y[yy])*(y1-y[yy]) ) ) )
    
    ondas.write("\n")
    
ondas.close()

data = loadtxt("ondas_estanque.dat", float)
imshow(data, origin="lower", extent=[0, 100, 0, 100])
xlabel("Distancia(cm)")
ylabel("Distancia(cm)")
zita1=zita*sin(k*r*(pi/180))
print("La función en la posición (x,y)",zita1)
show()
```

#EJERCICIO: Interferencia de ondas 

Gráfica de las ondas generadas al caer dos piedras con una separación de 20 cm 

```python
from numpy import zeros, array, loadtxt
from math import pi, sqrt, sin
from pylab import imshow, xlabel,ylabel, plot, show
puntos=500
dimen=100. #cm
paso=dimen/puntos #cm/puntos
lmda=5.0 #cm
k=2*pi/lmda
zita=1.0 #cm
x1=float(input("Ingrese la posición en x: "))
y1=float(input("Ingrese la posición en y: "))

x=zeros(puntos, float)
y=zeros(puntos, float)

for i in range(puntos):
    x[i]=paso*i #creo vectores
    y[i]=paso*i
    
ondas=open("ondas_estanque.dat", "w")

for yy in range( len(y) ):
    for xx in range( len(x) ):
        r = sqrt((x1-x[xx])*(x1-x[xx]) + (y1-y[yy])*(y1-y[yy]) )
        r1= sqrt(((x1+20)-x[xx])*((x1+20)-x[xx]) + (y1-y[yy])*(y1-y[yy]) )
        zita1=zita*sin(k*r)
        zita2=zita*sin(k*r1)
        #ondas.write("%0.1f " % (zita1+zita2))
        ondas.write("%0.1f " % (zita1))
        ondas.write("%0.1f " % (zita2))
    
    ondas.write("\n")
    
ondas.close()

data = loadtxt("ondas_estanque.dat", float)
imshow(data, origin="lower", extent=[0, 100, 0, 100])
xlabel("Distancia en x (cm)")
ylabel("Distancia en y (cm)")
show()
```
